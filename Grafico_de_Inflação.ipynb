```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados de inflação (IPCA) anual
anos = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
ipca = [10.67, 6.29, 2.95, 3.75, 4.31, 4.52, 10.06, 5.79, 4.62]

df = pd.DataFrame({
    'Ano': anos,
    'Inflação (IPCA)': ipca
})

plt.figure(figsize=(10, 5))
plt.plot(df['Ano'], df['Inflação (IPCA)'], marker='o', color='mediumvioletred', linewidth=2)
plt.title('Inflação no Brasil (IPCA) - 2015 a 2023')
plt.xlabel('Ano')
plt.ylabel('Inflação (%)')
plt.grid(True)
plt.xticks(anos)
plt.ylim(0, max(ipca) + 2)

for i in range(len(df)):
    plt.text(df['Ano'][i], df['Inflação (IPCA)'][i] + 0.3, f"{df['Inflação (IPCA)'][i]}%", ha='center')

plt.tight_layout()
plt.show()
```
